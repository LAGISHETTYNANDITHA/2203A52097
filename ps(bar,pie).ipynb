```python
import pandas as pd
d=pd.read_csv("/content/netflix_titles.csv")
print(d)
```

```python
s=(d.groupby('country').size().to_frame('number of movies').reset_index())
s=s[(s['number of movies']>=80)]
print(s)
```

```python
plt.figure(figsize=(20,7))
plot=plt.bar(s['country'],s['number of movies'],width=0.6,color=['red'])
for i in plot:
  h=i.get_height()
  w=i.get_width()
  x=i.get_x()
plt.text(x+w/2.5,0.5*h,int(h))
plt.xlabel('country')
plt.xticks(rotation=90)
plt.ylabel('no of movies released')
plt.show()
```

```python
values=[42,50,32,55,9,12]
mylabels=['lung','breast','colon','prostate','melanoma','bladder']
mycolors=['red','yellow','orange','cyan','green','pink']
myexplode=[0,0.5,0,0,0,0]
stangle=90
plt.pie(values,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors,startangle=90,autopct='%1.1f%%')
```
