```python
import pandas as pd
import numpy as np
d = pd.read_csv('/content/insurance.csv')
print(d)
```

```python
string_columns = ['sex', 'smoker', 'region']
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for col in string_columns:
    le = LabelEncoder()
    d[col] = le.fit_transform(d[col])
    label_encoders[col] = le
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(d['age'], d['expenses'], c='red', marker='+', s=80)

for i in range(len(d)):
    if i % 50 == 0:
        plt.text(
            d['age'][i],
            d['expenses'][i],
            f"({d['age'][i]}, {d['expenses'][i]})",
            fontsize=8
        )

# Customizing the plot
plt.title('Scatter Plot of Age vs. Expenses in Insurance Dataset', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Expenses', fontsize=14)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
filtered_data = d[d['expenses'] > 20000]

plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['age'], filtered_data['expenses'], c='red', marker='+')
plt.title('Scatter Plot of Age vs Expenses (Filtered)')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.show()
```

```python
a = d[d['age'] > 39]['smoker'].value_counts()
b = d[d['age'] <= 39]['smoker'].value_counts()

plt.figure(figsize=(8, 6))

plt.bar(['Above 39', 'Below 39'], [a.sum(), b.sum()], color=['skyblue', 'lightcoral'])

plt.xlabel("Age Group")
plt.ylabel("Number of Smokers")
plt.title("Number of Smokers Above and Below 39")

plt.show()
```
